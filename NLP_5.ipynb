## Strojno uenje (eng. machine learning)
je dio umjetne inteligencije pomou kojega se nastoje razviti algoritmi koji bi
imali sposobnost samostalnog uenja.

* **nadzirano uenje (eng. supervised learning)**:
  algoritmi ue na temelju oznaenih podataka, obuhvaa klasifikaciju i regresiju.,
  
* **nenadzirano uenje (eng. unsupervised learning)**: algoritmi ne ue na temelju oznaenih podataka, algoritmi nastoje grupirati ulazne podatke u klastere (za koje ne znaju to znae) ili pronai
uzorke (eng. patterns) u ulaznim podacima.


* **polunadzirano uenje (eng. semi-supervised learning)**:  kombinacija nadziranog i nenadziranog uenja.,

  
* **pojaano uenje (eng. reinforcement learning)**: temelji se na sustavu nagrada i kazni, koristi se u vie-agentskim sustavima.

### Algoritam k-sredina

- Algoritam k-sredina (eng. k-Means Algorithm, k-Means clustering) spada u algoritme
nenadziranog strojnog uenja (eng. unsupervised machine learning) i njegov cilj je pronai
uzorke koji postoje u odreenim podacima.
- Broj k u nazivu algoritma odnosi se na broj klastera na koje algoritam pokuava podijeliti
ulazne podatke.
- Algoritam ne zna to koji klaster znai (da zna ne bi spadao u postupke nenadziranog
uenja), zna samo da ih je sveukupno k i da mora podatke nekako raspodijeliti u tih k
klastera.










**Algoritam k-sredina++ (eng. k-means++ algorithm)** 
radi na slijedei nain:

* Nasumino se odabere sredite prvog poetnog klastera.
* Za svaku toku (tj. podatak) koju imamo rauna se
udaljenost do najblieg sredita klastera.
* Toka koja je najudaljenija najbliem sreditu klastera
postaje sredite novog klastera.
* Prethodna dva koraka se ponavljaju dok ne pronaemo k
sredita klastera.

**Metoda lakta (eng. elbow method)**:

* Kod ove metode variramo broj klastera od 1 do 10.
* U svakoj iteraciji metode raunamo zbroj kvadriranih udaljenosti naih
podataka (npr. piksela) od sredita klastera (koji se odreuju
nasumino). Ovaj se postupak zove zbroj kvadrata unutar klastera
(eng. Within-Cluster Sum of Squares, WCSS).
* Na temelju dobivenih rezultata nacrtamo graf iji je oblik slian obliku
lakta ljudske ruke.
* Na grafu pronaemo mjesto gdje on naglo postaje sve paralelniji sa xosi. To mjesto odgovara broju klastera koje moemo koristiti u algoritmu
k-sredina.


### Problemi kod reprezentacije dokumenata pomou jednojedininog kodiranja i rjenika izgraenog na setu podataka za treniranje algoritma:
* postoji velika vjerojatnost da e se u dokumentima za testiranje algoritma
pojaviti rijei koje nee biti u rjeniku izgraenom tijekom treniranja,
* nije optimalno za svaki dokument graditi veinom prazni vektor.
* Da bi se ovi problemi rijeili, esto se koristi **TF-IDF** (eng. Term
Frequency  Inverse Document Frequency) metoda pomou koje se
iz dokumenata vade znaajke (eng. features) koje se onda koriste za
reprezentaciju dokumenata.

### Algoritam oekivanja-maksimizacije

* Algoritam oekivanja-maksimizacije (eng. Expectation-Maximization Algorithm, EM
algorithm) spada u nenadzirane (eng. unsupervised) optimizacijske algoritme i
koristi se za procjenu najvee vjerojatnosti (eng. maximum likelihood estimation)
kada u bazi znanja postoje rupe, tj. latentne varijable koje nisu zabiljeene (tzv.
latent variables).
* Ovaj algoritam se sastoji od dva koraka: procjenjivanja latentnih varijabli (E-korak,
oekivanje) i optimizacije parametara modela za procjenu najvee vjerojatnosti (Mkorak, maksimizacija). Ova dva koraka se ponavljaju sve dok algoritam ne
konvergira.
* Ovaj se algoritam, kao i algoritam k-sredina, koristi za **grupiranje (klasteriranje)
podataka.**
* **Na primjer**, ako imamo skupinu starih dokumenata i znamo da su ih napisala dvojica
autora, ali ne znamo tko je napisao koji dokument, moemo koristiti EM algoritam da
grupira dokumente u dvije skupine na temelju odreenih znaajki.

Da bismo jednostavnije objasnili kako radi algoritam oekivanjamaksimizacije, potrebno je biti upoznat sa **binomnom
distribucijom (eng. binomial distribution).** - pr na linku
