## Strojno učenje (eng. machine learning)
je dio umjetne inteligencije pomoću kojega se nastoje razviti algoritmi koji bi
imali sposobnost samostalnog učenja.

* **nadzirano učenje (eng. supervised learning)**:
  algoritmi uče na temelju označenih podataka, obuhvaća klasifikaciju i regresiju.,
  
* **nenadzirano učenje (eng. unsupervised learning)**: algoritmi ne uče na temelju označenih podataka, algoritmi nastoje grupirati ulazne podatke u klastere (za koje ne znaju što znače) ili pronaći
uzorke (eng. patterns) u ulaznim podacima.


* **polunadzirano učenje (eng. semi-supervised learning)**: – kombinacija nadziranog i nenadziranog učenja.,

  
* **pojačano učenje (eng. reinforcement learning)**: temelji se na sustavu nagrada i kazni, koristi se u više-agentskim sustavima.

### Algoritam k-sredina

- Algoritam k-sredina (eng. k-Means Algorithm, k-Means clustering) spada u algoritme
nenadziranog strojnog učenja (eng. unsupervised machine learning) i njegov cilj je pronaći
uzorke koji postoje u određenim podacima.
- Broj k u nazivu algoritma odnosi se na broj klastera na koje algoritam pokušava podijeliti
ulazne podatke.
- Algoritam ne zna što koji klaster znači (da zna ne bi spadao u postupke nenadziranog
učenja), zna samo da ih je sveukupno k i da mora podatke nekako raspodijeliti u tih k
klastera.










**Algoritam k-sredina++ (eng. k-means++ algorithm)** 
radi na slijedeći način:

* Nasumično se odabere središte prvog početnog klastera.
* Za svaku točku (tj. podatak) koju imamo računa se
udaljenost do najbližeg središta klastera.
* Točka koja je najudaljenija najbližem središtu klastera
postaje središte novog klastera.
* Prethodna dva koraka se ponavljaju dok ne pronađemo k
središta klastera.

**Metoda lakta (eng. elbow method)**:

* Kod ove metode variramo broj klastera od 1 do 10.
* U svakoj iteraciji metode računamo zbroj kvadriranih udaljenosti naših
podataka (npr. piksela) od središta klastera (koji se određuju
nasumično). Ovaj se postupak zove zbroj kvadrata unutar klastera
(eng. Within-Cluster Sum of Squares, WCSS).
* Na temelju dobivenih rezultata nacrtamo graf čiji je oblik sličan obliku
lakta ljudske ruke.
* Na grafu pronađemo mjesto gdje on naglo postaje sve paralelniji sa xosi. To mjesto odgovara broju klastera koje možemo koristiti u algoritmu
k-sredina.


### Problemi kod reprezentacije dokumenata pomoću jednojediničnog kodiranja i rječnika izgrađenog na setu podataka za treniranje algoritma:
* postoji velika vjerojatnost da će se u dokumentima za testiranje algoritma
pojaviti riječi koje neće biti u rječniku izgrađenom tijekom treniranja,
* nije optimalno za svaki dokument graditi većinom prazni vektor.
* Da bi se ovi problemi riješili, često se koristi **TF-IDF** (eng. Term
Frequency – Inverse Document Frequency) metoda pomoću koje se
iz dokumenata vade značajke (eng. features) koje se onda koriste za
reprezentaciju dokumenata.

### Algoritam očekivanja-maksimizacije

* Algoritam očekivanja-maksimizacije (eng. Expectation-Maximization Algorithm, EM
algorithm) spada u nenadzirane (eng. unsupervised) optimizacijske algoritme i
koristi se za procjenu najveće vjerojatnosti (eng. maximum likelihood estimation)
kada u bazi znanja postoje “rupe”, tj. latentne varijable koje nisu zabilježene (tzv.
latent variables).
* Ovaj algoritam se sastoji od dva koraka: procjenjivanja latentnih varijabli (E-korak,
“očekivanje”) i optimizacije parametara modela za procjenu najveće vjerojatnosti (Mkorak, “maksimizacija”). Ova dva koraka se ponavljaju sve dok algoritam ne
konvergira.
* Ovaj se algoritam, kao i algoritam k-sredina, koristi za **grupiranje (klasteriranje)
podataka.**
* **Na primjer**, ako imamo skupinu starih dokumenata i znamo da su ih napisala dvojica
autora, ali ne znamo tko je napisao koji dokument, možemo koristiti EM algoritam da
grupira dokumente u dvije skupine na temelju određenih značajki.

Da bismo jednostavnije objasnili kako radi algoritam očekivanjamaksimizacije, potrebno je biti upoznat sa **binomnom
distribucijom (eng. binomial distribution).** - pr na linku
