- Analiza osjeaja (eng. sentiment analysis) i detekcija emocija (eng. emotion detection) su pojmovi koji se esto koriste u obradi prirodnog jezika.

- Analiza osjeaja iz teksta koristi se za svrstavanje teksta u
(najee) jednu od tri klase: pozitivnu, negativnu ili neutralnu

- Analiza emocija iz teksta koristi se za pokuaj detekcije ljudskih emocija kao to su ljutnja, gaenje, strah, srea, tuga i
iznenaenje.


### Analiza osjeaja i detekcija emocija imaju brojne primjene u obradi prirodnog jezika:
 detekcija neeljene elektronike pote,
 predvianje rezultata politikih izbora,
 analiza popularnosti javnih osoba,
 detekcija lanih vijesti (eng. fake news),
 detekcija emocionalnog stanja osoba iz poruka na drutvenim mreama,
 raunanje indeksa sree (eng. happiness index) raznih zemalja 

### Postoje dva najea modela emocija koja se koriste u obradi prirodnog jezika:
 univerzalne emocije (Paul Ekman), i cirkularni model emocija (James. A. Russell).

 Ekmanov model univerzalnih emocija spada u diskretne modele budui
da emocije dijeli na 6 (ponekad 7) strogo definiranih klasa koje se ne preklapaju.

 Russellov cirkularni model emocija pretpostavlja da su emocije
povezane i da prema tome ne mogu biti diskretne.

#### SentiWordNet biblioteka u Pythonu za odreenu rije vraa postotak
povezanosti te rijei sa 3 klase: pozitivno, negativno i neutralno
(objektivno). mozemo vidjeti ispod primjer koristenja biblioteke


```python
import nltk
from nltk.corpus import sentiwordnet as swn
from nltk.corpus import wordnet as wn

nltk.download('wordnet') # WordNet
nltk.download('sentiwordnet')
```

```python
synsets = wn.synsets("happy")
print(synsets)

prvi_synset = synsets[0]

print("Prvi synset: " + str(synsets[0]))
print("Definicija: " + prvi_synset.definition())

swn_synset = swn.senti_synset(prvi_synset.name())

pozitivniPostotak = swn_synset.pos_score()
negativniPostotak = swn_synset.neg_score()
objektivniPostotak = swn_synset.obj_score()

print("Pozitivni postotak: " + str(pozitivniPostotak))
print("Negativni postotak: " + str(negativniPostotak))
print("Objektivni postotak: " + str(objektivniPostotak))
```

```python
synsets = wn.synsets("goat")

goat_syn = synsets[0];

swn_synset = swn.senti_synset(goat_syn.name())

pozitivniPostotak = swn_synset.pos_score()
negativniPostotak = swn_synset.neg_score()
objektivniPostotak = swn_synset.obj_score()

print(goat_syn)
print("Definicija: " + goat_syn.definition())
print("Pozitivni postotak: " + str(pozitivniPostotak))
print("Negativni postotak: " + str(negativniPostotak))
print("Objektivni postotak: " + str(objektivniPostotak))
```

```python
from senticnet.senticnet import SenticNet

sn = SenticNet()
rijec = "anger"

concept_info = sn.concept(rijec)

polarity_label = sn.polarity_label(rijec)
polarity_value = sn.polarity_value(rijec)

moodtags = sn.moodtags(rijec)
semantics = sn.semantics(rijec)
sentics = sn.sentics(rijec)

print(concept_info)
```
