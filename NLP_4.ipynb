- Analiza osjećaja (eng. sentiment analysis) i detekcija emocija (eng. emotion detection) su pojmovi koji se često koriste u obradi prirodnog jezika.

- Analiza osjećaja iz teksta koristi se za svrstavanje teksta u
(najčešće) jednu od tri klase: pozitivnu, negativnu ili neutralnu

- Analiza emocija iz teksta koristi se za pokušaj detekcije ljudskih emocija kao što su ljutnja, gađenje, strah, sreća, tuga i
iznenađenje.


### Analiza osjećaja i detekcija emocija imaju brojne primjene u obradi prirodnog jezika:
– detekcija neželjene elektroničke pošte,
– predviđanje rezultata političkih izbora,
– analiza popularnosti javnih osoba,
– detekcija lažnih vijesti (eng. fake news),
– detekcija emocionalnog stanja osoba iz poruka na društvenim mrežama,
– računanje indeksa sreće (eng. happiness index) raznih zemalja 

### Postoje dva najčešća modela emocija koja se koriste u obradi prirodnog jezika:
– univerzalne emocije (Paul Ekman), i cirkularni model emocija (James. A. Russell).

● Ekmanov model univerzalnih emocija spada u diskretne modele budući
da emocije dijeli na 6 (ponekad 7) strogo definiranih klasa koje se ne preklapaju.

● Russellov cirkularni model emocija pretpostavlja da su emocije
povezane i da prema tome ne mogu biti diskretne.

#### SentiWordNet biblioteka u Pythonu za određenu riječ vraća postotak
povezanosti te riječi sa 3 klase: pozitivno, negativno i neutralno
(objektivno). mozemo vidjeti ispod primjer koristenja biblioteke


```python
import nltk
from nltk.corpus import sentiwordnet as swn
from nltk.corpus import wordnet as wn

nltk.download('wordnet') # WordNet
nltk.download('sentiwordnet')
```

```python
synsets = wn.synsets("happy")
print(synsets)

prvi_synset = synsets[0]

print("Prvi synset: " + str(synsets[0]))
print("Definicija: " + prvi_synset.definition())

swn_synset = swn.senti_synset(prvi_synset.name())

pozitivniPostotak = swn_synset.pos_score()
negativniPostotak = swn_synset.neg_score()
objektivniPostotak = swn_synset.obj_score()

print("Pozitivni postotak: " + str(pozitivniPostotak))
print("Negativni postotak: " + str(negativniPostotak))
print("Objektivni postotak: " + str(objektivniPostotak))
```

```python
synsets = wn.synsets("goat")

goat_syn = synsets[0];

swn_synset = swn.senti_synset(goat_syn.name())

pozitivniPostotak = swn_synset.pos_score()
negativniPostotak = swn_synset.neg_score()
objektivniPostotak = swn_synset.obj_score()

print(goat_syn)
print("Definicija: " + goat_syn.definition())
print("Pozitivni postotak: " + str(pozitivniPostotak))
print("Negativni postotak: " + str(negativniPostotak))
print("Objektivni postotak: " + str(objektivniPostotak))
```

* Prema [3], oznake u SenticNet bazi su slijedećeg polariteta:
– sensitivity (anger/fear),
– attitude (trust/disgust),
– introspection (joy/sadness),
– temper (anticipation/surprise).

```python
from senticnet.senticnet import SenticNet

sn = SenticNet()
rijec = "anger"

concept_info = sn.concept(rijec)

polarity_label = sn.polarity_label(rijec)
polarity_value = sn.polarity_value(rijec)

moodtags = sn.moodtags(rijec)
semantics = sn.semantics(rijec)
sentics = sn.sentics(rijec)

print(concept_info)
```
