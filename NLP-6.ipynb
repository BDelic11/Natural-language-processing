### Linearna regresija

**Linearna regresija (eng. linear regression)** je linearni model koji se u statistici i
umjetnoj inteligenciji koristi za predvianje kontinuiranih vrijednosti te koji
pretpostavlja da postoji linearna veza izmeu ulaznih varijabli i izlazne
varijable.
* Jednostavnije, linearnom regresijom se trae veze meu podacima, tj.
pokuava se odrediti koliko je koja ulazna varijabla vana za eljeni izlaz.
* Linearna regresija se dijeli na:
 **jednostavnu linearnu regresiju** (eng. simple linear regression) gdje se na temelju
samo jedne ulazne varijable pokuava odrediti izlazna varijabla, te
 **viestruku linearnu regresiju** (eng. multiple linear regression) gdje se na temelju vie
ulaznih varijabli pokuava odrediti izlazna varijabla

* Linearna regresija se u obradi prirodnog jezika koristi za rudarenje teksta
(eng. text mining, text data mining).
* Primjeri obrade prirodnog jezika u kojima se moe koristiti linearna
regresija:
 * automatsko odreivanje prosjene ocjene (npr. broja zvjezdica) odreene knjige,
filma ili serije na temelju tekstualnih recenzija,
 * automatsko ocjenjivanje eseja u kolama,
 * automatska procjena psihikog stanja osobe na temelju tekstualnih statusa na
drutvenim mreama,
 * automatsko procjenjivanje cijene zdravstvenog osiguranja osobe na temelju
tekstualnih zapisa o zdravlju te osobe, njezinim dobrim i loim ivotnim navikama,
mjestu prebivalita, itd.



### Generativni i diskriminativni modeli

* Generativni modeli (eng. generative models) iz ulaznih podataka ue
zajednike vjerojatnosti (eng. joint probabilities) (npr. kolika je vjerojatnost da
se dogaaji x i y dogode zajedno, tj. rauna se p(x,y)).
* Diskriminativni modeli (eng. discriminative models) iz ulaznih podataka ue
uvjetne vjerojatnosti (eng. conditional probabilities) (npr. koja je vjerojatnost
da se dogodi dogaaj x ako se je dogodio dogaaj y, tj. rauna se p(x|y)).
* Primjer: elimo napraviti algoritam koji bi mogao prepoznati jezik na kojemu je
tekst napisan.
* Generativni model: naui sve mogue jezike (tj. vjerojatnosnu distribuciju koja
pokazuje kako se podaci generiraju) i zatim pokua klasificirati tekst u ispravnu klasu.
* Diskriminativni model: naui razlike izmeu svih moguih moguih jezika i zatim
pokua klasificirati tekst u ispravnu klasu.

 **Primjeri generativnih modela:**
 
* naivni Bayesov klasifikator,
* skriveni Markovljevi modeli.
* Primjeri diskriminativnih modela:
* logistika regresija,
* SVM (algoritam stroja potpornih vektora, eng. Support Vector Machines),
* umjetne neuralne mree,
 stabla odluka.

### Naivni Bayesov klasifikator

* Naivni Bayesov klasifikator (eng. Nave Bayes classifier) spada u kategoriju
generativnih modela i probabilistikih klasifikatora.
* Temelji se na primjenjivaju Bayesovog teorema:
* Od svih klasifikatora koji postoje, a koji su utemeljeni na Bayesovom teoremu,
naivni Bayesov klasifikator je najjednostavniji.
* Rije naivni u imenu ovoga klasifikatora dolazi od naivne pretpostavke da su
znaajke ili podaci koji se koriste prilikom klasifikacije neovisni jedni o drugima.
U obradi prirodnog jezika ovo esto znai da klasifikator zanemaruje redoslijed
rijei u reenicama, tj. zanemaruje se kontekst u kojem se rije pojavljuje.

**P(AB)=P(BA)P(A)/ P(B)**

* P (A|B)  uvjetna vjerojatnost ili posteriorna vjerojatnost (eng. posterior
probability)  vjerojatnost da se dogodi dogaaj A ako se je dogodio dogaaj
B.
* P (B|A)  uvjetna vjerojatnost ili likelihood  vjerojatnost da se dogodi dogaaj
B ako se je dogodio dogaaj A. Jo se zove i likelihood od A budui da za
fiksni B vrijedi P(B|A) = L(A|B).
* P(A), P(B)  vjerojatnost da se dogodi dogaaj A, odnosno dogaaj B. Jo se
zove i granina vjerojatnost (eng. marginal probability) ili prethodna
vjerojatnost (eng. prior probability).
