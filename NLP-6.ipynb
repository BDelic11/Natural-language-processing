### Linearna regresija

**Linearna regresija (eng. linear regression)** je linearni model koji se u statistici i
umjetnoj inteligenciji koristi za predviđanje kontinuiranih vrijednosti te koji
pretpostavlja da postoji linearna veza između ulaznih varijabli i izlazne
varijable.
* Jednostavnije, linearnom regresijom se traže veze među podacima, tj.
pokušava se odrediti koliko je koja “ulazna” varijabla važna za željeni izlaz.
* Linearna regresija se dijeli na:
– **jednostavnu linearnu regresiju** (eng. simple linear regression) gdje se na temelju
samo jedne ulazne varijable pokušava odrediti izlazna varijabla, te
– **višestruku linearnu regresiju** (eng. multiple linear regression) gdje se na temelju više
ulaznih varijabli pokušava odrediti izlazna varijabla

* Linearna regresija se u obradi prirodnog jezika koristi za rudarenje teksta
(eng. text mining, text data mining).
* Primjeri obrade prirodnog jezika u kojima se može koristiti linearna
regresija:
 * automatsko određivanje prosječne ocjene (npr. broja zvjezdica) određene knjige,
filma ili serije na temelju tekstualnih recenzija,
 * automatsko ocjenjivanje eseja u školama,
 * automatska procjena psihičkog stanja osobe na temelju tekstualnih statusa na
društvenim mrežama,
 * automatsko procjenjivanje cijene zdravstvenog osiguranja osobe na temelju
tekstualnih zapisa o zdravlju te osobe, njezinim dobrim i lošim životnim navikama,
mjestu prebivališta, itd.



### Generativni i diskriminativni modeli

* Generativni modeli (eng. generative models) iz ulaznih podataka uče
zajedničke vjerojatnosti (eng. joint probabilities) (npr. kolika je vjerojatnost da
se događaji x i y dogode zajedno, tj. računa se p(x,y)).
* Diskriminativni modeli (eng. discriminative models) iz ulaznih podataka uče
uvjetne vjerojatnosti (eng. conditional probabilities) (npr. koja je vjerojatnost
da se dogodi događaj x ako se je dogodio događaj y, tj. računa se p(x|y)).
* Primjer: želimo napraviti algoritam koji bi mogao prepoznati jezik na kojemu je
tekst napisan.
* Generativni model: nauči sve moguće jezike (tj. vjerojatnosnu distribuciju koja
pokazuje kako se podaci generiraju) i zatim pokuša klasificirati tekst u ispravnu klasu.
* Diskriminativni model: nauči razlike između svih mogućih mogućih jezika i zatim
pokuša klasificirati tekst u ispravnu klasu.

 **Primjeri generativnih modela:**
 
* naivni Bayesov klasifikator,
* skriveni Markovljevi modeli.
* Primjeri diskriminativnih modela:
* logistička regresija,
* SVM (algoritam stroja potpornih vektora, eng. Support Vector Machines),
* umjetne neuralne mreže,
– stabla odluka.

### Naivni Bayesov klasifikator

* Naivni Bayesov klasifikator (eng. Naïve Bayes classifier) spada u kategoriju
generativnih modela i probabilističkih klasifikatora.
* Temelji se na primjenjivaju Bayesovog teorema:
* Od svih klasifikatora koji postoje, a koji su utemeljeni na Bayesovom teoremu,
naivni Bayesov klasifikator je najjednostavniji.
* Riječ “naivni” u imenu ovoga klasifikatora dolazi od naivne pretpostavke da su
značajke ili podaci koji se koriste prilikom klasifikacije neovisni jedni o drugima.
U obradi prirodnog jezika ovo često znači da klasifikator zanemaruje redoslijed
riječi u rečenicama, tj. zanemaruje se kontekst u kojem se riječ pojavljuje.

**P(A∣B)=P(B∣A)P(A)/ P(B)**

* P (A|B) – uvjetna vjerojatnost ili posteriorna vjerojatnost (eng. posterior
probability) – vjerojatnost da se dogodi događaj A ako se je dogodio događaj
B.
* P (B|A) – uvjetna vjerojatnost ili likelihood – vjerojatnost da se dogodi događaj
B ako se je dogodio događaj A. Još se zove i likelihood od A budući da za
fiksni B vrijedi P(B|A) = L(A|B).
* P(A), P(B) – vjerojatnost da se dogodi događaj A, odnosno događaj B. Još se
zove i granična vjerojatnost (eng. marginal probability) ili prethodna
vjerojatnost (eng. prior probability).
